- edits = no. of edits
- size = total size of edits
- users = no. of contributing users

```python
import pandas as pd
import numpy as np
```

```python
fn_out = '../output/wikipedia_metrics.csv'
```

```python
fn = '../input/wikipedia_edits.csv.zip'
df = pd.read_csv(fn)
```

```python
df.head()
```

# Aggregate Using GroupBy

```python
df_edits = pd.DataFrame(df.groupby("title").size())
df_edits.head()
```

```python
df_edits.columns = ['edits']
```

```python
df_edits.head()
```

```python
df_edits.sort_values(by='edits', ascending=False).head()
```

```python
df_sizes = pd.DataFrame(df.groupby('title')['size'].sum())
df_sizes.head()
```

```python
df.groupby('title')['size'].agg([len, np.sum, np.mean, np.median, np.std])
```

```python
df_sizes.sort_values(by='size')
```

```python
df_users = pd.DataFrame(df.groupby('title')['user'].nunique())
df_users.sort_values(by='user', ascending=False)
```

```python
df_edits.reset_index(inplace=True)
df_sizes.reset_index(inplace=True)
df_users.reset_index(inplace=True)
```

```python
df_out = pd.merge(df_edits, df_sizes)
df_out = pd.merge(df_out, df_users)
df_out.head()
```

```python
df_out.to_csv(fn_out, index=False)
```
